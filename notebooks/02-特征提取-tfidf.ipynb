```python
import pickle
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus=pd.read_csv("../data/train_set.csv", sep="\t")["text"].values
print("{} train data".format(len(corpus)))
corpus=np.concatenate([corpus, pd.read_csv("../data/test_a.csv", sep="\t")["text"].values], axis=0)
print("{} all data".format(len(corpus)))
```

```python
corpus[0]
```

```python
tfidfv = TfidfVectorizer(analyzer="word", min_df=3, max_features=5000, use_idf=True)
def train_tfidf(corpus):
    out=tfidfv.fit_transform(corpus)
    print("shape={}".format(out.toarray().shape))
    with open("../models/tfidf-model.pkl", "wb") as f:
        pickle.dump(tfidfv, f)

train_tfidf(corpus)
```
